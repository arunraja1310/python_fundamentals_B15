```python
#List Datatype in python
my_list=['apple','orange','banana','grapes','mosambi','pineapple','papaya']
print(my_list)
```

```python
#list indexing 
# indexing is easy for list traversal
#easy to identify the position of element in a list
for i in range(len(my_list)):
    print('index of\t'+str(my_list[i])+'\tis='+str(i))
```

```python
#add ,delete and replace a elements in list:
# add new element in the existing list
my_list.append('watermelon')#1.will always add element at end of the list
print(my_list)
```

```python
my_list.insert(0,'sappota') #2.while we add element at desire position we use insert 
print(my_list)
```

```python
#replace a element in list'
my_list[9]='lichi' #we use only index values to replace an element
print(my_list)
```

```python
#delete a element in a list
del my_list[7] #delete element using index
print(my_list)
```

```python
print(my_list.pop())#pop will only remove the element at end of list
```
