```python
#F strings in string data type

my_fname='arun'
my_lname='rajan'
my_name=f"{my_fname}{my_lname}"
print(my_name)
```

```python
#Enhancement of 'F' string
my_name.title()
```

```python
#find data type of the variable
type(my_name)
```

```python
my_name.upper() #to display all letters in uppercase
```

```python
my_name.lower() #to display all letters in lowercase
```

```python
#whitespace in string 
myString='hello '
print(myString.lstrip()) #lstrip to remove leftside whitespace
```

```python
myString=' welcome'
print(myString.rstrip())#rstrip to remove rightside whitespace
```

```python
myString=' thankyou '
print(myString.strip())#strip used to remove either sideof whitespaces
```

```python
#adding white space in string we use delimiters
#1.newline delimiter
myString='welcome\nthankyou'
print(myString)
```

```python
#2.tabspace delimiter
myString="Home\twork"
print(myString)
```
